```python
import seaborn as sns
import pandas as pd
df = pd.read_csv('Diwali Sales Data.csv',encoding='unicode_escape')
ax= sns.countplot(data=df,x='Zone',hue='Gender')
```

```python
df.columns
```

```python
sales_state= df.groupby(['State'], as_index=False )['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(17,5)})
sns.barplot(data=sales_state,x='State',y='Orders')
```

```python
sales_amount = df.groupby(['State'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_amount , x='State' , y='Amount')
```

```python
ax=sns.countplot(data=df , x='Marital_Status')
sns.set(rc={'figure.figsize':(7,5)})
# for bar in ax.containers:
#     ax.bar_label(bars)
```

```python
sale_state= df.groupby(['Marital_Status','Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount').head(10)
sns.set(rc={'figure.figsize':(10,5)})
sns.barplot(data=sale_state, x='Marital_Status',y='Amount',hue='Gender')
```

```python
ax = sns.countplot(data = df , x='Occupation')
sns.set(rc={'figure.figsize':(30,5)})
```

```python
sales_occ = df.groupby(['Occupation'],as_index=False )['Amount'].sum().sort_values(by='Amount',ascending=False).head(7)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data= sales_occ , x='Occupation',y='Amount')
```

```python
df.columns
```

```python
pro = sns.countplot(data=df,x='Product_Category')
sns.set(rc={'figure.figsize':(50,5)})
```

```python
ab = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(7)
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data=ab, x='Product_Category',y='Amount')
```

```python
sales_state = df.groupby(['Product_ID'],as_index=False)['Orders'].sum().sort_values(by=['Orders'],ascending=False).head(7)
sns.set({'figure.figsize':(15,5)})
sns.barplot(data= sales_state , x='Product_ID',y='Orders')
```
